```python
from selenium import webdriver as wb

from selenium.webdriver.common.keys import Keys

from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup as bs

from selenium.webdriver.chrome.service import Service

from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
```

```python
def find(driver,j):
    try:
        but=driver.find_elements(By.CLASS_NAME,'poster')[j]
        #driver.execute_script("arguments[0].scrollIntoView(true);", but)
        but.click() 
        time.sleep(2)
        soup = bs(driver.page_source, 'lxml')
        driver.back()
        title=soup.select_one('h2.title-kr').text.strip()
        genre = soup.select('span.item__body')[0].text.strip()
        air_date = soup.select('span.item__body')[2].text.strip().split(' ')[0]
        channel = soup.select('span.item__body')[4].text
        if air_date == '' or 'KBS' in channel or 'MBC' in channel:
            return -1
        list= [title,air_date,genre,channel]
        print(title)
        time.sleep(2)
    except:
        for i in range(9):
            body.send_keys(Keys.DOWN)
        time.sleep(2)
        j-=1
        return -1
    return list
```

```python
s = Service(ChromeDriverManager().install())
driver = wb.Chrome(service = s)  #
driver.get('https://m.kinolights.com/discover/explore')
```

```python
list=[] 
body = driver.find_element(By.CSS_SELECTOR,'body')
soup = bs(driver.page_source, 'lxml')

for j in range(200):
    f = find(driver,j)
    if f == -1 :
        j -= 1
    else:
        list.append(f)
```
